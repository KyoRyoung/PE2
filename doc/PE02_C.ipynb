## 1. Info

This software is programmed to analyze measured .xml data of semiconductor wafers. You can select the file you want or view the entire data analyzed. Through this program, you can view and save the processed data as a .png file or save the data as a .csv file.

- `python 3.9`
- `Windows 11`

## 3. Instructions and result

1. Run `run.py`
2. Select the desired device type (ex. `LMZC`, `LMZO`, `LMZ`)
3. Enter the desired wafer number in the form "D##" (ex. `D07`, `D08`, `D23`, `D24`, `all`)
4. Enter the date of measurement (`20YYMMDD`, `all`)
5. Enter Wafer's row and column in the form of "#,#" (ex. `0,0`)

6. Enter your name
7. Enter y/n whether to see the data as .png file or not
8. Enter y/n whether to save the data as .png file or not
9. Enter y/n whether to save the data as .csv file or not
10. Exit the program

## 4. Analysis

Below is an introduction to various wafer data analysis based on raw data.


```python
import os
import glob
import pandas as pd


matching_files = glob.glob("**/Analysis_Result_*.xlsx", recursive=True)

for file_path in matching_files:
    data = pd.read_excel(file_path)
data
```

```python
ref_spec_error_rows = data[data['Error Description'] == 'Ref. spec. Error']

# Check if all rows have no error
if len(ref_spec_error_rows) == 0:
    print("Error: None")

# Iterate over the filtered rows and display the required information
for index, row in ref_spec_error_rows.iterrows():
    print("Wafer:", row['Wafer'])
    print("Date:", row['Date'])
    print("Row:", row['Row'])
    print("Column:", row['Column'])
    print("")
```
