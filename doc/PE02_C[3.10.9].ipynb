````python
import os
import glob
import pandas as pd
from IPython.display import Image, display

matching_files = glob.glob("../res/Analysis_Result_*.xlsx", recursive=True)

data_list = [] 

for file_path in matching_files:
    data = pd.read_excel(file_path)
    data_list.append(data)

combined_data = pd.concat(data_list, ignore_index=True)

# Error Description에 Ref.spec.Error가 있는지 확인한다.
has_error = combined_data['Error Description'].str.contains('Ref.spec.Error')

if has_error.any():
    filtered_data = combined_data[has_error]

    for index, row in filtered_data.iterrows():
        lot = row['Lot']
        wafer = row['Wafer']
        mask = row['Mask']
        test_site = row['TestSite']
        row_value = row['Row']
        column_value = row['Column']

        png_name = f"{lot}_{wafer}_({row_value},{column_value})_{mask}_{test_site}.png"
        #만일 에러가 있으면 에러가 있는 사진을 찾아 그래프를 보여준다.
        png_path = glob.glob(f"../**/{wafer}/**/{png_name}", recursive=True)
        
        
        print("Ref.spec.Error")

        if png_path:
            try:
                display(Image(filename=png_path[0]))
            except Exception as e:
                print(f"Error displaying PNG file: {e}")
        else:
            print("Error: PNG file not found")
else:
    print("No Error")
````

````python
import matplotlib.pyplot as plt
import numpy as np
print ('Red dot is Ref.spec.Error')

grouped_data_testsite = data.groupby('TestSite')


for testsite, group_testsite in grouped_data_testsite:

    grouped_data_wafer = group_testsite.groupby('Wafer')

    # TestSite마다 Wafer를 나눈다.
    for wafer, group_wafer in grouped_data_wafer:
 
        fig, ax = plt.subplots(figsize=(6, 3))  

        x = group_wafer['Row']
        y = group_wafer['Column']
        error_description = group_wafer['Error Description']

        # Error Description에서 Rer.spec.Error이면 빨강색, No Error이면 파란색 점이 되도록 한다.
        colors = ['red' if desc == 'Ref.spec.Error' else 'blue' for desc in error_description]


        ax.scatter(x, y, c=colors, s=50)  

  
        ax.grid(True, linestyle='--')


        ax.set_xlabel('Row')
        ax.set_ylabel('Column')
        ax.set_title(f'TestSite: {testsite}, Wafer: {wafer}')

        # x축 y축범위 설정
        x_buffer = max(abs(x.min()), abs(x.max())) + 1
        y_buffer = max(abs(y.min()), abs(y.max())) + 1
        ax.set_xlim(-x_buffer, x_buffer)
        ax.set_ylim(-y_buffer, y_buffer)


        ax.xaxis.set_major_locator(plt.MultipleLocator(1))
        ax.yaxis.set_major_locator(plt.MultipleLocator(1))


        ax.axhline(0, color='black', linewidth=1.5)
        ax.axvline(0, color='black', linewidth=1.5)


        plt.show()

````

````python
import matplotlib.pyplot as plt

# Lot 별로 data를 나눈다.
grouped_data = data.groupby('Lot')


for lot, group in grouped_data:

    fig, axs = plt.subplots(1, 2, figsize=(12, 6))
    fig.suptitle(f'Lot {lot}', fontsize=16)

    # Wavelength 1550nm인 데이터를 찾는다
    wavelength_1550 = group[group['Analysis Wavelength(nm)'] == 1550]
    wafer_1550 = wavelength_1550['Wafer']

    # Wavelength 1310nm인 데이터를 찾는다
    wavelength_1310 = group[group['Analysis Wavelength(nm)'] == 1310]
    wafer_1310 = wavelength_1310['Wafer']

    # 각 wavelength에 대하여 갯수를 센다.
    count_1550 = wafer_1550.value_counts()
    count_1310 = wafer_1310.value_counts()

    # 1550nm에 대한 그래프
    axs[0].bar(count_1550.index, count_1550, label='1550nm', alpha=0.7)
    axs[0].set_title('1550nm')
    axs[0].set_xlabel('Wafer')
    axs[0].set_ylabel('Count')

   
    for i, count in enumerate(count_1550):
        axs[0].text(count_1550.index[i], count - 1, str(count), ha='center', va='bottom')

    # 1310nm에 대한 그래프
    axs[1].bar(count_1310.index, count_1310, label='1310nm', alpha=0.7)
    axs[1].set_title('1310nm')
    axs[1].set_xlabel('Wafer')
    axs[1].set_ylabel('Count')

    
    for i, count in enumerate(count_1310):
        axs[1].text(count_1310.index[i], count - 1, str(count), ha='center', va='bottom')

    
    plt.show()
````

#### 4-2-2. Rsq of Ref.spectrum

````python
import matplotlib.pyplot as plt

# Lot 별 Rsq of Ref.spectrum 평균 계산
lot_avg_rsq = data.groupby(['Lot', 'Wafer'])['Rsq of Ref.spectrum(Nth)'].mean()

# 웨이퍼별 평균 계산
wafer_avg_rsq = data.groupby('Wafer')['Rsq of Ref.spectrum(Nth)'].mean()

# 전체 Lot의 평균 계산
overall_avg_rsq = lot_avg_rsq.mean()


fig, ax = plt.subplots(figsize=(12, 8))

# 웨이퍼별 Rsq of Ref.spectrum 평균 그래프
lot_avg_rsq.unstack('Lot').plot(kind='bar', stacked=True, ax=ax)


ax.axhline(overall_avg_rsq, color='red', linestyle='--', label='Overall Avg.')
ax.text(len(lot_avg_rsq), overall_avg_rsq, f"Overall Avg.: {overall_avg_rsq:.4f}", ha='center', va='bottom')

plt.title('Average Rsq of Ref.spectrum by Wafer (Lot-wise)')
plt.xlabel('Wafer')
plt.ylabel('Average Rsq')
plt.legend(title='Lot')


plt.xticks(ticks=range(len(lot_avg_rsq.index.levels[1])), labels=lot_avg_rsq.index.levels[1], rotation=0, ha='center')
plt.gca().set_aspect('equal')


for i, avg_rsq in enumerate(wafer_avg_rsq):
    ax.text(i, avg_rsq, f"{avg_rsq:.4f}", ha='center', va='bottom')

plt.tight_layout()
plt.show()

````

#### 4-2-3. Max transmission

````python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# TestSite별로 데이터를 나눈다.
grouped_data = data.groupby('TestSite')


for test_site, group in grouped_data:
    # 각 TestSite에서 Wafer별로 나눈다.
    wafer_grouped = group.groupby('Wafer')


    num_wafers = len(wafer_grouped)
    num_cols = 2
    num_rows = (num_wafers + 1) // 2


    fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 6*num_rows), sharex=True)
    fig.suptitle(f'Transmission Distribution - TestSite: {test_site}', fontsize=16)


    axes = axes.flatten()


    for i, (wafer, wafer_data) in enumerate(wafer_grouped):
        # Wafer에서 Max transmission을 찾는다.
        transmission = wafer_data['Max transmission of Ref spec. (dB)']

        # kernel density estimation (KDE)를 구한다.
        kde = np.random.normal(0, 0.02, len(transmission))
        density, bins, _ = axes[i].hist(kde, bins=30, alpha=0)  # Calculate density without plotting

        # 밀도의 최대 찾기
        max_density_bin = bins[np.argmax(density)]

        
        y_values = np.random.uniform(-10, -4, len(transmission))
        axes[i].scatter(np.zeros_like(transmission), y_values, alpha=0.5)

        # 평균과 분산 계산
        mean = np.mean(transmission)
        variance = np.var(transmission)

        # 신뢰구간 95%를 찾는다.
        confidence_interval = norm.interval(0.95, loc=mean, scale=np.sqrt(variance))
        axes[i].axhline(mean, color='red', linestyle='--', label='Mean')
        axes[i].axhspan(confidence_interval[0], confidence_interval[1], facecolor='gray', alpha=0.3, label='95% Confidence Interval')
        annotation = f"Start: {confidence_interval[0]:.4f}\nEnd: {confidence_interval[1]:.4f}\nMean: {mean:.4f}\nVariance: {variance:.4f}"
        axes[i].annotate(annotation, (0, mean), xytext=(10, 10), textcoords='offset points', ha='left', va='bottom')
        axes[i].set_ylim(-10, -4)       
        axes[i].set_ylabel('Transmission')
        axes[i].set_title(f'Wafer {wafer}')
 
    if num_wafers % 2 != 0:
        fig.delaxes(axes[-1])

    axes[-1].set_xlabel('')
    axes[0].legend()
    fig.tight_layout()


plt.show()

````
